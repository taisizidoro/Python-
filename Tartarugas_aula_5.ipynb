Sabendo o ângulo interno da ponta de uma estrela, desenhe quatro estrelas em uma janela, com uma certa distância entre elas. Dica: use a função penup() e pendown() do módulo turtle.

```
def varios_poligonos():

  num_de_lados = input(print("Quantos lados tem?"))
  ang = (360/(2*(int(num_de_lados))))
  comprimento = 70
  num_de_pol = input(print("quantas formas serão?"))

  lais = t
  lais.initializeTurtle()

  for u in range(int(num_de_pol)):
    for i in range(int(num_de_lados)):
      lais.forward(comprimento)
      lais.left(180 - ang)
    lais.penup()
    lais.left(360/(int(num_de_pol)))
    lais.forward(2*comprimento)
    lais.pendown()

varios_poligonos()
```

Escreva um conjunto apropriadamente geral de funções que possam desenhar flores:

```
import math

carla = t
carla.initializeTurtle()
carla.speed(10)
carla.color("purple")
carla.width(1)

def polygon(turtle, lenght, n, percent):
  alfa = 360*percent/n
  for i in range(n):
    turtle.forward(lenght)
    turtle.left(alfa)

def circle(turtle, r, percent = 1):
  circunference = 2*math.pi*r
  n = 10
  lenght = int(circunference/n)
  polygon(turtle, lenght, n, percent)

def petal(percent):
  r = 10
  angle = 180 - percent*360

  circle(carla, r, percent)
  carla.left(angle)
  circle(carla, r, percent)

#7 petalas 
carla.penup()
carla.left(90)
carla.forward(200)
carla.pendown()
for i in range(7):
  petal(1/5)
  carla.left(56.5)

#10 pétalas 
carla.penup()
carla.left(180)
carla.forward(200)
carla.pendown()
for i in range(10):
  petal(1/5)
  carla.left(360/5)

#20 pétalas 
carla.penup()
carla.forward(200)
carla.pendown()
for i in range(20):
  petal(1/5) #pra abrir as petalas é só colocar 1/10
  carla.left(90)

petal(1/5)
```

Escreva um conjunto apropriadamente geral de funções que podem desenhar formas como na figura: (tortas)

```
import math 

aline = t
aline.initializeTurtle()
aline.speed(10)

#para ir pro canto 
aline.penup()
aline.left(90)
aline.forward(250)
aline.right(90)
aline.pendown()
#terminou aqui

def torta(lados):
  tamanho = 100
  ang1 = 360/lados
  ang2 = (180 - ang1)/2
  raio = tamanho*(math.sin(ang2*math.pi/180)/math.sin(ang1*math.pi/180))

  for i in range(lados):
    aline.forward(int(raio))
    aline.left(180-ang2)
    aline.forward(tamanho)
    aline.left(180-ang2)
    aline.forward(int(raio))
    aline.left(180)

  aline.penup()
  aline.right(90)
  aline.forward(250)
  aline.left(90)
  aline.pendown()

torta(5)
torta(6)
torta(7)
```

As letras do alfabeto podem ser construídas a partir de um número moderado de elementos básicos, como linhas verticais e horizontais e algumas curvas. Projete um alfabeto que possa ser desenhado com um número mínimo de elementos básicos e depois escreva funções que desenhem as letras.
Você deve escrever uma função para cada letra, com os nomes draw_a_, _draw_b_, etc., e colocar suas funções em um arquivo chamado letters.py.

```

lele = t
lele.initializeTurtle()
lele.color("red")
lele.width(1)

def arc(turtle, r, percent = 1):
  circunference = 2*math.pi*r
  n = 10
  lenght = int(circunference/(2*n))
  polygon(turtle, lenght, n, percent)

def f(t, length):    #vai para frente
    t.frente(length)

def tr(t, length):    #vai para trás
    t.tras(length)

def lt(t, angle=90): #vai para esquerda
    t.left(angle)

def rt(t, angle=90): #vai para direita
    t.right(angle)

def pd(t):            #levanta a caneta
    t.pendown()

def pu(t):            #abaixa a caneta
    t.penup()

##############


def fl(t, n, angle=90):     #frente e esquerda
    f(t, n)
    lt(t, angle)

def ft(t, n):               #frente e trás
    f(t, n)
    bk(t, n)

def skip(t, n):             #pula certo comprimento
    pu(t)
    f(t, n)
    pd(t)

def linha_rt(t, n, angle=90):  #faz uma linha e vira para direita
    lt(t)
    f(t, n)
    rt(t, angle)

def pula_rt(t, n):           #move a turtle pra frente e vira pra direita
    lt(t)
    skip(t, n)
    rt(t)

#############


def post(t, n):           #frente e trás e vira pra direita
    lt(t)
    ft(t, n)
    rt(t)

def beam(t, n, height):   #pula, faz uma linha e volta pra posição inicial
    pula_rt(t, n*height)
    ft(t, n)
    pula_rt(t, -n*height)

def vt(t, n, height):  #linha vertical e horizontal e volta
    linha_rt(t, n * height)
    ft(t, n)
    lt(t)
    tr(t, n*height)
    rt(t)

def diagonal(t, x, y):      #faz uma diagonal
    from math import atan2, sqrt, pi
    angle = atan2(y, x) * 180 / pi
    dist = sqrt(x**2 + y**2)
    lt(t, angle)
    ft(t, dist)
    rt(t, angle)

def v(t, n, height):     #faz forma de v
    diagonal(t, -n/2, height*n)
    diagonal(t, n/2, height*n)

def bump(t, n, height):
    linha_rt(t, n*height)
    arc(t, n/2.0, 180)
    lt(t)
    ft(t, n*height+n)

############# AGORA AS LETRAS ###################


def draw_a(t, n):
    diagonal(t, n/2, 2*n)
    beam(t, n, 1)
    skip(t, n)
    diagonal(t, -n/2, 2*n)

def draw_b(t, n):
    bump(t, n, 1)
    bump(t, n, 0)
    skip(t, n/2)

def draw_c(t, n):
    vt(t, n, 2)
    f(t, n)

def draw_d(t, n):
    bump(t, 2*n, 0)
    skip(t, n)

def draw_e_f(t, n):
    vt(t, n, 2)
    vt(t, n, 1)

def draw_e(t, n):
    draw_e_f(t, n)
    ft(t, n)

def draw_f(t, n):
    draw_e_f(t, n)
    skip(t, n)

def draw_g(t, n):
    vt(t, n, 2)
    f(t, n/2)
    beam(t, n/2, 2)
    f(t, n/2)
    post(t, n)

def draw_h(t, n):
    post(t, 2*n)
    vt(t, n, 1)
    skip(t, n)
    post(t, 2*n)

def draw_i(t, n):
    beam(t, n, 2)
    f(t, n/2)
    post(t, 2*n)
    f(t, n/2)

def draw_j(t, n):
    beam(t, n, 2)
    arc(t, n/2, 90)
    f(t, 3*n/2)
    skip(t, -2*n)
    rt(t)
    skip(t, n/2)

def draw_k(t, n):
    post(t, 2*n)
    stump(t, n, 180)
    v(t, 2*n, 0.5)
    ft(t, n)
    skip(t, n)

def draw_l(t, n):
    post(t, 2*n)
    f(t, n)

def draw_n(t, n):
    post(t, 2*n)
    skip(t, n)
    diagonal(t, -n, 2*n)
    post(t, 2*n)

def draw_m(t, n):
    post(t, 2*n)
    draw_v(t, n)
    post(t, 2*n)

def draw_o(t, n):
    skip(t, n)
    circle(t, n)
    skip(t, n)

def draw_p(t, n):
    bump(t, n, 1)
    skip(t, n/2)

def draw_q(t, n):
    draw_o(t, n)
    diagonal(t, -n/2, n)

def draw_r(t, n):
    draw_p(t, n)
    diagonal(t, -n/2, n)

def draw_s(t, n):
    f(t, n/2)
    arc(t, n/2, 180)
    arc(t, n/2, -180)
    ft(t, n/2, -90)
    skip(t, 2*n)
    lt(t)

def draw_t(t, n):
    beam(t, n, 2)
    skip(t, n/2)
    post(t, 2*n)
    skip(t, n/2)

def draw_u(t, n):
    post(t, 2*n)
    f(t, n)
    post(t, 2*n)

def draw_v(t, n):
    skip(t, n/2)
    v(t, n, 2)
    skip(t, n/2)

def draw_w(t, n):
    draw_v(t, n)
    draw_v(t, n)

def draw_x(t, n):
    diagonal(t, n, 2*n)
    skip(t, n)
    diagonal(t, -n, 2*n)

def draw_v(t, n):
    skip(t, n/2)
    diagonal(t, -n/2, 2*n)
    diagonal(t, n/2, 2*n)
    skip(t, n/2)

def draw_y(t, n):
    skip(t, n/2)
    stump(t, n)
    v(t, n, 1)
    rt(t)
    ft(t, n)
    skip(t, n/2)

def draw_z(t, n):
    beam(t, n, 2)
    diagonal(t, n, 2*n)
    f(t, n)

def draw_(t, n):    #dá um espaço
    skip(t, n)

#########################

tmnh = 20

for f in [draw_b, draw_o, draw_m, draw_, draw_d, draw_i, draw_a]:
    f(lele, tmnh)
    skip(lele, tmnh)
```
