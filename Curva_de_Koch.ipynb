Para desenhar uma curva de Koch com o comprimento x, tudo o que você tem que fazer é:

Desenhe uma curva de Koch com o comprimento x/3.

Vire 60 graus à esquerda.

Desenhe uma curva de Koch com o comprimento x/3.

Vire 120 graus à direita.

Desenhe uma curva de Koch com o comprimento x/3.

Vire 60 graus à esquerda.

Desenhe uma curva de Koch com o comprimento x/3.

A exceção é se x for menor que 3: neste caso, você pode desenhar apenas uma linha reta com o comprimento x.

Escreva uma função chamada koch que receba um ColabTurtle (ou turtle) e um comprimento como parâmetros, e use o ColabTurtle para desenhar uma curva de Koch com o comprimento dado.

Escreva uma função chamada snowflake que desenhe três curvas de Koch para fazer o traçado de um floco de neve.

A curva de Koch pode ser generalizada de vários modos. Veja exemplos em http://en.wikipedia.org/wiki/Koch_snowflake e implemente o seu favorito.

```
!pip3 install ColabTurtle
import ColabTurtle
import ColabTurtle.Turtle as t
from ColabTurtle import *
```

```
def koch(j, x):       #desenha uma curva de koch de tamanho x
    if x < 10:
      j.forward(int(x))
      return
    m = x/3 
    koch(j, m)        #desenha a curva de comprimento x/3
    j.left(60)        #vira 
    koch(j, m)        #desenha outra curva
    j.right(120)      #vira
    koch(j, m)        #....
    j.left(60)
    koch(j, m)


def snowflake(j, n):        #aqui desenha o floquinho com as curvas
    for i in range(3):
        koch(j, n)
        j.right(120)

def floco():              #aqui inicializa a turtle e chama a função do floquinho
  bob = t
  bob.initializeTurtle()
  bob.speed(10)
  bob.width(1)
  bob.penup()
  bob.pendown()
  snowflake(bob, 300)

floco()
```
