Quantos intervalos (ou qual largura de intervalo) é preciso implementar para ter uma precisão de 1% no resultado numérico na integração de $f(x)$ de 0 a 10?

$\dfrac{R_{num}-R_{an}}{R_{an}} \leq 1\% $

```
import matplotlib.pyplot as plt
import numpy as np
import math

#valores da questão: x_i = 0, x_f = 10, dx = 1, precisão = 1
def Riemann_precisao():
 
  x_i = int(input(print("Entre com o valor inicial:")))
  x_f = int(input(print("Entre com o valor final:")))
  dx = int(input(print("Entre com o valor de dx:")))
  precisão = int(input(print("Entre com o valor da precisão:")))
  S_analitica = math.exp(x_i)-math.exp(-x_f)

  while precisão >= 0.01: 
    n_steps = round((x_f-x_i)/dx)
    x = np.arange(x_i, x_f, dx)
    S_numerica = 0 
    for xi in x: 
      S_numerica = S_numerica + math.exp(-xi)*dx
    precisão = abs(S_numerica-S_analitica)/S_analitica
    print("Aproximação:", S_numerica, "/", "Número de caixas:", n_steps, "/", "Erro:", precisão*100, "%")
    print()
    dx = dx*0.1
  
Riemann_precisao()
```

Faça a integração numérica pela esquerda da mesma função. Compare o resultado com a integração pela direita.

```
import matplotlib.pyplot as plt
import numpy as np
import math

#valores da questão: x_i = 0, x_f = 10, dx = 1
def f(x):
  return math.exp(-x)

def Riemann_esquerda(f,x_i,x_f,dx):
  x = np.arange(x_i, x_f, dx)
  S_numerica = 0 
  for i in range(len(x)-1): 
    S_numerica = S_numerica + f(x[i])*dx
  print("Soma de Riemann pela esquerda:", S_numerica)

  x=range(0,10)
  y=[math.exp(-xi) for xi in x]
  plt.bar(x,y,color="red",align="edge",width=1)
  x1 = np.linspace(0,10, 1000) 
  y1=[math.exp(-xi) for xi in x1]
  plt.plot(x1,y1)
  plt.show()

def Riemann_direita(f,x_i,x_f,dx):
  x = np.arange(x_i, x_f, dx)
  S_numerica = 0 
  for i in range(1,len(x)): 
    S_numerica = S_numerica + f(x[i])*dx
  print("Soma de Riemann pela direita:", S_numerica)

  x=range(0,10)
  y=[math.exp(-xi) for xi in x]
  plt.bar(x,y,color="red",align="edge",width=1)
  x1 = np.linspace(0,10, 1000) 
  y1=[(math.exp(-xi+1)) for xi in x1]
  plt.plot(x1,y1)
  plt.show()
  
Riemann_direita(f,0,10,1)
Riemann_esquerda(f,0,10,1)
```

Faça a integração por trapézios. Lembre que a área de um trapézio é dada por: $A=\frac{(a + b)h}{2}$

```
import numpy as np
import math

def f(x):
  return math.exp(-x)

def Riemann_trapezio(f,x_i,x_f,dx):
  x = np.arange(x_i, x_f, dx)
  S_numerica = 0 
  for i in range(len(x)-dx): 
    trap = (((f(x[i])+(f(x[i+dx])))*dx)/2)
    S_numerica = S_numerica + trap*dx
  print("Soma de Riemann pelo trapézio:", S_numerica)

  x=range(0,10)
  y=[math.exp(-xi) for xi in x]
  plt.fill_between(x,y,color="red")
  x1 = np.linspace(0,10, 1000) 
  y1=[math.exp(-xi) for xi in x1]
  plt.plot(x1,y1)
  plt.show()

Riemann_trapezio(f,0,10,1)
```

Faça a integração numérica das funções trigonométricas $\sin(x)$ e $\cos(x)$ num período 0 (menor) x (menor ou igual) 2*pi

```
import numpy as np
import math

#usei 2pi como aproximadamente 6 para ser inteiro 

def f1(x):
  return math.sin(x)

def f2(x):
  return math.cos(x)

def Riemann_seno(f,x_i,x_f, dx):
  x = np.arange(x_i, x_f, dx)
  S_numerica = 0 
  for i in x:
    sen = f1(x[i])
    S_numerica = S_numerica + sen*dx
  print("Soma de Riemann para o seno:", S_numerica)

  x=range(0,7)
  y=[math.sin(xi) for xi in x]
  plt.fill_between(x,y,color="red")
  x1 = np.linspace(0,6, 1000) 
  y1=[math.sin(xi) for xi in x1]
  plt.plot(x1,y1)
  plt.show()

def Riemann_cosseno(f,x_i,x_f,dx):
  x = np.arange(x_i, x_f, dx)
  S_numerica = 0 
  for i in x:
    cos = f2(x[i])
    S_numerica = S_numerica + cos*dx
  print("Soma de Riemann para o cosseno:", S_numerica)

  x=range(0,7)
  y=[math.cos(xi) for xi in x]
  plt.fill_between(x,y,color="red")
  x1 = np.linspace(0,6, 1000) 
  y1=[math.cos(xi) for xi in x1]
  plt.plot(x1,y1)
  plt.show()

Riemann_seno(f1,0, 6, 1)
Riemann_cosseno(f2,0, 6, 1)
```
